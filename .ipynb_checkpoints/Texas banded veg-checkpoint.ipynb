```python
import numpy as np
import matplotlib.pylab as plt
```

```python
fname = '../Texas_Veg_Patterns/tx2010g.txt'
f = open(fname, 'r') # 'r' = read
# for line in f:
#   print line, # note, coma erases the "cartridge return"
# f.close()

ncols = int(f.next().strip().split()[1])
nrows = int(f.next().strip().split()[1])
xllcorner = int(f.next().strip().split()[1])
yllcorner = int(f.next().strip().split()[1])
cellsize = int(f.next().strip().split()[1])
NODATA_value = int(f.next().strip().split()[1])
print ncols,',', nrows
```

```python
nrows = 5000
ncols = ncols - 130
```

```python
ncols
```

```python
for j in range(100):
    f.next()
img = np.zeros([nrows, ncols])
for j in range(nrows):
    img[j,:] = np.array([float(l) for l in f.next().strip().split()])[:-130]
    
```

```python
img2 = np.zeros([nrows, ncols])
for j in range(nrows):
    img2[j,:] = np.array([float(l) for l in f.next().strip().split()])[:-130]    
```

```python
img2[img2 ==-9999.0] =  np.nan
```

```python
plt.figure(figsize = (12, 12))
plt.imshow(img, interpolation = 'none')
```
